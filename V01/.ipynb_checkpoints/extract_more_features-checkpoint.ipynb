```python
# reading the output from MOBO
constraint_model = pd.read_csv('constraint_model0.csv', index_col=[0])

# create extra variables 
def categorize_gyield (df, target, threshold):
    yield_cat = []
    for i in range(0, len(df)):
        if df[target].iloc[i] <= threshold:
            yield_cat.append(0)
        else:
            yield_cat.append(1)
    target_cat = target + '_Cat'
    df[target_cat] = yield_cat
    return df[target_cat]

threshold_GYield = 52
constraint_model ['GYield_Cat'] = categorize_gyield (df=constraint_model, 
                                                     target='GYield', 
                                                     threshold=threshold_GYield)
threshold_Yield = 82
constraint_model ['Yield_Cat'] =  categorize_gyield (df=constraint_model, 
                                                     target='Yield', 
                                                     threshold=threshold_Yield)
constraint_model ['charge_density'] = constraint_model ['cap']*\
                                      constraint_model ['volt']/\
                                      constraint_model ['mass']
print ('# of samples Yield  <= {}% in group 0:{}'.format (threshold_Yield, 
                                                       sum (constraint_model['Yield_Cat']==0)))
print ('# of samples Yield  >  {}% in group 1:{}'.format (threshold_Yield, 
                                                       sum (constraint_model['Yield_Cat']==1)))
print ('# of samples GYield <= {}% in group 0:{}'.format (threshold_GYield, 
                                                       sum (constraint_model['GYield_Cat']==0)))
print ('# of samples GYield >  {}% in group 1:{}'.format (threshold_GYield, 
                                                       sum (constraint_model['GYield_Cat']==1)))

constraint_model ['I_Max_norm'] = np.array (constraint_model['I_Max']/constraint_model['mass'])
constraint_model ['charge_tot_norm'] = constraint_model['charge_tot']/(constraint_model['mass'])*1000
constraint_model ['I_Final_perc'] = np.array (constraint_model['I_Final']/constraint_model['I_Max'])
constraint_model.columns
```

```python
# proxy (current-time plot) prediction (Final_Current_percent)
features_proxy = np.concatenate ((
    np.array (constraint_model ['charge_density']).reshape(-1, 1),
    np.array (constraint_model ['ptime']).reshape(-1, 1),
    np.array (constraint_model ['R1_CB']).reshape(-1, 1),
    np.array (constraint_model ['R1_MC']).reshape(-1, 1),
    np.array (constraint_model ['R1_PA']).reshape(-1, 1),
    np.array (constraint_model ['pre_volt']).reshape(-1, 1),
    np.array (constraint_model ['pre_peak']).reshape(-1, 1)), axis=1)

feature_proxy_list = ['charge_denisty', 'PulseTime', 'material_CB',
                'material_MC', 'material_PA', 'Pretreat_voltage', 'Pretreat_peak']

rand=9
indices = np.arange(len(features_proxy))

train_features_proxy, test_features_proxy, idx_train_proxy, idx_test_proxy =\
               train_test_split(features_proxy, indices, test_size=0.2, random_state = rand)
```

```python
# training proxy models only training data
# predicting I_Final_perc 
I_Final_perc = np.array (constraint_model['I_Final_perc'])
train_I_Final_perc = I_Final_perc[idx_train_proxy]
test_I_Final_perc = I_Final_perc[idx_test_proxy]
# fixed hyperparameters
xgbr_I_Final_perc = XGBRegressor(max_depth=2, n_estimators=46, gamma=0.01, 
                                     learning_rate=0.115, subsample=0.75, 
                                     min_child_weight=2, random_state=0)
# Train the model on all data
xgbr_I_Final_perc.fit(train_features_proxy, train_I_Final_perc);
I_Final_perc_predicted = xgbr_I_Final_perc.predict (features_proxy)
pred_I_Final = xgbr_I_Final_perc.predict (test_features_proxy)
print ('I_Final accuracy', r2_score(test_I_Final_perc, pred_I_Final))

# predicting I_Max_norm 
I_Max_norm = np.array (constraint_model['I_Max_norm'])
train_I_Max_norm = I_Max_norm[idx_train_proxy]
test_I_Max_norm = I_Max_norm[idx_test_proxy]
# fixed hyperparameters
xgbr_I_Max_norm = XGBRegressor(max_depth=6, n_estimators=10, gamma=0.001, 
                   learning_rate=0.15, subsample=0.77, min_child_weight=5, 
                   random_state=0)
# Train the model on all data
xgbr_I_Max_norm.fit(train_features_proxy, train_I_Max_norm);
I_Max_norm_predicted = xgbr_I_Max_norm.predict (features_proxy)
pred_I_Max = xgbr_I_Max_norm.predict (test_features_proxy)
print ('I_Max accuracy', r2_score(test_I_Max_norm, pred_I_Max))

# predicting charge_tot_norm
charge_tot_norm = np.array (constraint_model['charge_tot_norm'])
train_charge_tot_norm = charge_tot_norm[idx_train_proxy]
test_charge_tot_norm = charge_tot_norm[idx_test_proxy]
# fixed hyperparameters
xgbr_charge_tot_norm = XGBRegressor(max_depth=5, n_estimators=14, gamma=0.001, 
                   learning_rate=0.149, subsample=0.8, min_child_weight=2, 
                  random_state=0)
# Train the model on all data
xgbr_charge_tot_norm.fit(train_features_proxy, train_charge_tot_norm);
charge_tot_norm_predicted = xgbr_charge_tot_norm.predict (features_proxy)
pred_charge_tot_norm = xgbr_charge_tot_norm.predict (test_features_proxy)
print ('charge_tot_norm accuracy', r2_score(test_charge_tot_norm, pred_charge_tot_norm))
```

```python
# Final GYield and Yield input features
features_final = np.concatenate ((
    np.array (constraint_model ['charge_density']).reshape(-1, 1),
    np.array (constraint_model ['ptime']).reshape(-1, 1),
    np.array (constraint_model ['R1_CB']).reshape(-1, 1),
    np.array (constraint_model ['R1_MC']).reshape(-1, 1),
    np.array (constraint_model ['R1_PA']).reshape(-1, 1),
    np.array (constraint_model ['pre_volt']).reshape(-1, 1),
    np.array (constraint_model ['pre_peak']).reshape(-1, 1),
    I_Final_perc_predicted.reshape(-1, 1),
    I_Max_norm_predicted.reshape(-1, 1),
    charge_tot_norm_predicted.reshape(-1, 1)), axis=1)

rand=0
indices = np.arange(len(features_final))
train_features_final, test_features_final, idx_train_final, idx_test_final = \
                 train_test_split(features_final, indices, test_size=0.2, random_state = rand)
feature_final_list = [
    'charge_denisty', 'PulseTime', 'material_CB', 'material_MC', 
    'material_PA', 'Pretreat_voltage', 'Pretreat_peak', 
    'I_Final_perc_predicted', 'I_Max_norm_predicted', 'charge_tot_norm_predicted']
```

```python
#final prediction
GYield_Cat = np.array (constraint_model ['GYield_Cat'])
Yield_Cat = np.array (constraint_model ['Yield_Cat'])
train_GYield_Cat, test_GYield_Cat = GYield_Cat[idx_train_final], GYield_Cat[idx_test_final]
train_Yield_Cat, test_Yield_Cat = Yield_Cat[idx_train_final], Yield_Cat[idx_test_final]

xgbc_gyield = XGBClassifier(
        max_depth=3, n_estimators=30, gamma=0.01, eval_metric='error',
        learning_rate=0.115, subsample=0.75, min_child_weight=3, 
        use_label_encoder=False, random_state=0)
# Train the model on training data
xgbc_gyield.fit(train_features_final, train_GYield_Cat.ravel());
predictions = xgbc_gyield.predict(features_final)
total_acc = accuracy_score(GYield_Cat, predictions)
print ('simple percentage of correct guesses: ', total_acc)

xgbc_yield = XGBClassifier(
        max_depth=2, use_label_encoder=False, n_estimators=50, gamma=0.01, eval_metric='error',
        learning_rate=0.1, subsample=0.75, min_child_weight=2, random_state=0)

# Train the model on training data
xgbc_yield.fit(features_final, Yield_Cat.ravel());
predictions = xgbc_yield.predict(features_final)
total_acc = accuracy_score(Yield_Cat, predictions)
print ('simple percentage of correct guesses: ', total_acc)
```

```python
# change the confidence, 
# reduce False positive (bad conditions do not pass the constraint)
pred_prob_yield = xgbc_yield.predict_proba(test_features_final)[:, 1]
pred_yield_60per = [0 if prob<=0.6 else 1 for prob in pred_prob_yield]
total_acc = accuracy_score(test_Yield_Cat, pred_yield_60per)
print (total_acc)

# Define classifiers
conversion_indices = {'aC':0, 'G':1}

# Plot confusion matrix
confusion_matrix = np.zeros((len(conversion_indices), len(conversion_indices)))

# First row is Graphene Yield < 50% == 0
# Second row is Graphene Yield >= 50% == 1
for i in range(len(test_Yield_Cat)):
    confusion_matrix[int(test_Yield_Cat[i]), int(pred_yield_60per[i])] += 1
    
print("\n===== Confusion Matrix (between subject's) =====")
print(confusion_matrix)

for i in range(confusion_matrix.shape[0]):
    # print(sum(confusion_matrix[i, :]))
    confusion_matrix[i, :] = confusion_matrix[i, :]/sum(confusion_matrix[i, :])
    
print("\n===== Confusion Matrix % (between subject's) =====")
print(confusion_matrix)
```
